```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python
def get_color(name, number):
    pal = list(sns.color_palette(palette=name, n_colors=number).as_hex())
    return pal
```

```python
df_2 = pd.read_excel("Все_источники_ЗПВ_2023.xlsx")
df_2.info()
```

```python
# преобразуем в сводную таблицу
df2=df_2.pivot_table(
    values='Количество записей на прием к врачу',
    index='target_nsi_organization_level1_short_name_dynamic',
    columns='appointment_chanel_name_custom_epgu',
    aggfunc='sum'
)
df_3=df2.fillna(0) # заменяем пустые значения 0
df_3
```

```python
df3 = df_3. reset_index() # преобразуем в DataFrame
df3
```

```python
# строим гистограмму
df2['Врач_АПУ'].plot(kind='bar',
    figsize=(20, 10),
    title='Врач_АПУ_ЗПВ_2023',
    grid = True
);
```

```python
####################################################### Гистограмма диаграмма ###############################################################
data_excel = "Врач_АПУ"
name_excel = "target_nsi_organization_level1_short_name_dynamic"
name = "Гистограмма ЗПВ_2023"
df = df3

# создаем отсортированный DataFrame
# df_filtered = df.sort_values(by=data_excel, ascending=False) #сортировка по убыванию

#задаем цвета
pal_vi = get_color('viridis_r', len(df))
# строим гистограмму
fig = px.bar(df, x=name_excel, y=data_excel, text=data_excel,
color =name_excel, color_discrete_sequence=pal_vi)

fig.update_traces(texttemplate='%{text:.3s}', textposition='outside')
fig.update_layout({'plot_bgcolor': 'white',
'paper_bgcolor': 'white',
'title': {'text': name, 'y':0.997}})  # Добавляем описание графика и сдвигаем его немного вниз

# центруем
fig.update_layout(width=1100, height=500,
margin = dict(t=15, l=15, r=15, b=15))

# выводим
fig.show()
fig.write_image("Гистограмма ЗПВ_2023.svg")
```

```python
####################################################### Накопительная гистограмма диаграмма ###############################################################
data_excel_1 = "Врач_АПУ"
data_excel_2 = "ЕПГУ"
name_excel = "target_nsi_organization_level1_short_name_dynamic"
name = "гистограмма записи пациентов через врача и ЕПГУtarget_nsi_organization_level1_short_name_dynamic"
df = df3

# Извлекаем данные из DataFrame
data1 = df[data_excel_1]
data2 = df[data_excel_2]
names = df[name_excel]

# Создаем бины для гистограммы
bins = range(len(names))

# Устанавливаем размер фигуры
plt.figure(figsize=(16, 8))  # делаем более высокое расширение картинки

# Рисуем накопительную гистограмму с использованием палитры цветов
plt.bar(names, data1, label=data_excel_1)
plt.bar(names, data2, bottom=data1, label=data_excel_2)

# Настраиваем оси и метки
plt.xlabel(name_excel)
plt.ylabel("Количество, ед.")
plt.title(name)
plt.legend([data_excel_1, data_excel_2])
# plt.legend("12")
# Настраиваем подписи горизонтальной оси под углом 45 градусов
plt.xticks(rotation=90)

# Используем tight_layout для автоматической подгонки элементов графика
plt.tight_layout()

# Отображаем график и сохраняем его
plt.savefig("гистограмма записи пациентов через врача и ЕПГУ.png", dpi=1500)
plt.show()
plt.close('all')
```

```python
####################################################### Гоночная круговая диаграмма ###############################################################
df = df3
data_excel = "Врач_АПУ"
name_excel = "target_nsi_organization_level1_short_name_dynamic"
name = "Гоночная круговая диаграмма по способу записи в мед.учреждение"


# Создаем список цветов в переменных
pal_plas = get_color('Spectral', len(df))

# установим размер фигуры
plt.gcf().set_size_inches(12, 8)
sns.set_style('darkgrid')

# Установим максимальное значение
max_val = max(df[data_excel]) * 2
ax = plt.subplot(projection='polar')

# Зададим внутренний график
ax.set_theta_zero_location('N')
ax.set_theta_direction(1)
ax.set_rlabel_position(0)
ax.set_thetagrids([], labels=[])
ax.set_rgrids(range(len(df)), labels=df[name_excel])

# Установим график с полярной проекцией
ax = plt.subplot(projection='polar')
for i in range(len(df)):
    ax.barh(i, list(df[data_excel])[i] * 2 * np.pi / max_val,
label=list(df[name_excel])[i], color=pal_plas[i])

# Уменьшим размер шрифта подписей
ax.set_rgrids(range(len(df)), labels=df[name_excel], fontsize=6)

# Добавим подпись с названием графика
ax.set_title(name, fontsize=12)

# размещаем легенду справа от графика
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.tight_layout()
plt.savefig("гоночная диаграмма.png")

plt.show()
```
